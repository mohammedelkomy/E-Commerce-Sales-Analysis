```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1Ô∏è‚É£ Download data and overview

```python
df = pd.read_csv("dataset.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Date'] # object :(
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='%Y-%m-%d')
df['Date'] # Datetime :) 
```

2Ô∏è‚É£ Sales Analysis 

* Total Sales
* Average order value

```python
df['Total_Amount'].agg(['sum','mean']).round()
```

üîπ Sales by product category

```python
sales_by_category = df.groupby('Product_Category')['Total_Amount'].sum().round().sort_values(ascending=False)
sales_by_category
```

üîπ Sales By Year 

```python
sales_by_year = df['Date'].dt.year.value_counts().sort_index()
sales_by_year
```

üîπ Top-selling products by quantity

```python
df.groupby('Product_Category')['Quantity'].sum().sort_values(ascending=False)
```

3Ô∏è‚É£ Discount Analysis 

```python
df['Discount_Rate'] = df['Discount_Amount'] / (df['Unit_Price'] * df['Quantity'])
```

```python
df.groupby('Product_Category')['Discount_Rate'].mean().sort_values(ascending=False)
```

4Ô∏è‚É£ Customer Analysis 

üîπ Sales by gender

```python
df.groupby('Gender')['Total_Amount'].count()
```

üîπ New customers vs. returning customers

```python
counts = df['Is_Returning_Customer'].value_counts() 
# return percentage
percentages = (counts/counts.sum())*100
counts , percentages
```

üîπ Average spending of the top five customers

```python
df.groupby('Customer_ID')['Total_Amount'].mean().sort_values(ascending=False).head(5)
```

üîπ Average spending of the top ten customers by Age

```python
df.groupby('Age')['Total_Amount'].mean().sort_values(ascending=False).head(10)
```

```python
bins = [18,25,35,45,55,65,75]
labels = ['18-25','26-35','36-45','46-55','56-65','66-75']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels,include_lowest=True)
```

```python
df.groupby('Age_Group')['Total_Amount'].mean()
```

5Ô∏è‚É£ City Analysis 

üîπ Total sales for each city

```python
df.groupby('City')['Total_Amount'].sum().sort_values(ascending=False)
```

üîπ Average customers ratings by city

```python
df.groupby('City')['Customer_Rating'].mean().sort_values(ascending=False).round(2)
```

6Ô∏è‚É£ Payment methods 

```python
df['Payment_Method'].value_counts()
```

üîπ Total and average sales by payment method

```python
df.groupby('Payment_Method')['Total_Amount'].agg(['sum','mean']).sort_values(by='mean',ascending=False)
```
